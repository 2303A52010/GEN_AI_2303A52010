```python
from scipy.optimize import linprog
import numpy as np
obj = [-6, -5]
A = [[1,1],[3,2]]
b = [5,12]
bnd = [(0,float("inf")),(0,float("inf"))]
opt =linprog(c=obj, A_ub=A, b_ub=b, bounds=bnd)
choco = np.round(opt.x)
print(choco[0],choco[1])
```
